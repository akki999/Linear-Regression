```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('airline cost new.csv')
```

```python
df.head()
```

## Gradient Descent Algorithm

```python
X = df['num_pass']
y = df['cost']
```

```python
m = 10
c = 1200
n = len(X)
L = 0.0001

epochs = 1000

for i in range(epochs):
    y_pred = m*X + c
    D_m = (-2/n)*sum(X*(y - y_pred))
    D_c = (-2/n)*sum(y - y_pred)
    m = m-L*D_m
    c = c-L*D_c
    
    if(i%100)==0:
        plt.scatter(X, y)
        plt.plot(X, y_pred)
        plt.show()
        
print('m:', m)
print('c:', c)
```

Epochs is no of times entire data is run

batch_size is no of observations read per iterations

iteration = epochs * n / batch_size
